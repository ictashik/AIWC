## 2. Variables and Data Types
Python supports different data types such as:
- **Integers** (whole numbers)
- **Floats** (decimal numbers)
- **Strings** (text)
- **Booleans** (True/False)
- **Lists, Tuples, and Dictionaries**

```python
x = 10  # Integer
y = 3.14  # Float
name = "Alice"  # String
is_student = True  # Boolean
print("Integer:", x)
print("Float:", y)
print("String:", name)
print("Boolean:", is_student)
```

### Lists and Dictionaries
Lists store multiple values, while dictionaries store key-value pairs.

```python
fruits = ["Apple", "Banana", "Cherry"]
person = {"name": "Alice", "age": 25, "city": "New York"}
print("List:", fruits)
print("Dictionary:", person)
```

## 3. Conditional Statements
Conditional statements allow decision-making in Python.

```python
age = 18
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

## 4. Loops
Loops are used to iterate over sequences.

```python
for i in range(5):
    print("Iteration:", i)
```

## 5. Functions
Functions help in code reusability.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

## 6. Introduction to NumPy
NumPy is a package for scientific computing in Python.

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
print("Array Mean:", np.mean(arr))
```

## 7. Plotting with Matplotlib
Matplotlib is a library used for plotting graphs in Python.

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Plot")
plt.legend()
plt.show()
```

## 8. Introduction to Pandas
Pandas is used for handling structured data like tables.

```python
import pandas as pd

data = {"Name": ["Alice", "Bob", "Charlie"], "Age": [25, 30, 35], "City": ["New York", "London", "Paris"]}
df = pd.DataFrame(data)
print(df)
```
